### Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np 
```

### Opening the File

```python
fp = open('E:\Chanakya\Markets in Smartgrids\Config\hrl-smart-grids\DDPG\logs\price.txt', 'r')
file = fp.readlines()
```

### Extracting Data

```python
prices = []
# For quantity file[-40000:][100].split(',')[1].split(' ')[1].split('[')[1]
for i in range(40000):
    if file[-40000:][i].split(',')[0].split(' ')[0].split('[')[1] != '':
        prices.append(float(file[-40000:][i].split(',')[0].split(' ')[0].split('[')[1]))
```

###

```python
bins = [i for i in np.linspace(-0.5,20.5,22)]
fig= plt.figure(figsize=(20,15)) 
plt.hist(prices, bins)
plt.xticks(bins)
plt.xlabel('Bins')
plt.ylabel('Number of times')
plt.title('Analysis')
```
