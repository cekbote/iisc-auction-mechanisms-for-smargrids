### Importing Libraries


```python
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib notebook
```

### Preprocessing

```python
# Reward Logger Path
reward_logger = 'C:/Chanakya/Projects/Auction Mechanism for Smartgrids/hrl-smart-grids-master/DDPG/logs/reward_logger.txt'

# Rewards and Energy Transfered to the Grid Initialization
num_of_agents = 5
no_of_iterations = []
reward = []
energy_grid = []

for i in range(num_of_agents):
    no_of_iterations.append([])
    reward.append([])
    energy_grid.append([])

# Rewards and Energy Transfered to the Grid
with open(reward_logger, 'r') as re:
    for line in re:
        if '---' in line:
            pass
        else:
            list_ = line.strip().split(' | ')
            if len(list_) > 1 :
                iter_no = float(list_[0].split(': ')[1])
                agent_no = int(list_[1].split(': ')[1])
                main_reward = float(list_[2].split(': ')[1])
                energy_tranfs = float(list_[4].split(': ')[1])
                
                no_of_iterations[agent_no].append(iter_no)
                reward[agent_no].append(main_reward)
                energy_grid[agent_no].append(energy_tranfs)
```

### Graphs

#### Rewards

```python
plt.rcParams["figure.figsize"] = (20,16)
for i in range(num_of_agents):
    plt.subplot(3, 2, i + 1)
    plt.plot(no_of_iterations[i][1:], reward[i][1:], label = 'Agent ' + str(i + 1))
    plt.title('Agent ' + str(i+1))
    plt.xlabel('Number of Iterations')
    plt.ylabel('Average Rewards')
    
```

#### Energy Transfered to and from the Main Grid

```python
plt.rcParams["figure.figsize"] = (20,16)
for i in range(num_of_agents):
    plt.subplot(3, 2, i + 1)
    plt.plot(no_of_iterations[i][1:], energy_grid[i][1:], label = 'Agent ' + str(i + 1))
    plt.title('Agent ' + str(i+1))
    plt.xlabel('Number of Iterations')
    plt.ylabel('Average Energy Transfered')
```
